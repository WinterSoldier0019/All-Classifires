```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
Raw_Data = pd.read_csv("kc_house_data.csv")
Raw_Data.head()
```

```python
Raw_Data.shape
```

```python
Raw_Data.describe()
```

```python
Raw_Data.describe(include="all")
```

```python
D1 = Raw_Data.drop("id",axis=1)
D1.head()
```

```python
D1.isnull().sum()
```

```python
sns.distplot(D1["price"])
```

```python
quant = D1["price"].quantile(0.99)
D2 = D1[D1["price"] < quant]
print(quant)
```

```python
sns.distplot(D2["price"])
```

```python
D2.reset_index(drop=True)
```

```python
D2.head()
```

```python
log_price = np.log(D2["price"])
D2["Log_Price"] = log_price
D2.head()
```

```python
D3 = D2.drop("price",axis=1)
D3.head()
```

```python
D3.columns.values
```

```python
cols = ['Log_Price','bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat',
       'long', 'sqft_living15', 'sqft_lot15']
```

```python
Final_data = D3[cols]
```

```python
Final_data.head()
```

```python
x = Final_data.iloc[:, 1:19]
x.head()
```

```python
y = Final_data["Log_Price"]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler as SS
```

```python
Scaler = SS()
Scaler.fit(x)
```

```python
x_scaled = Scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=19)
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
Reg = LR()
Reg.fit(x_train,y_train)
```

```python
Reg.score(x_train,y_train)
```

```python
Reg.score(x_test,y_test)
```

```python
y_hat = Reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel("Actual",fontsize=20)
plt.ylabel("Predicted",fontsize=20)
plt.legend()
plt.show()
```

```python
plt.scatter(y_train,y_hat,alpha=0.3)
plt.xlabel("Actual",fontsize=20)
plt.ylabel("Predicted",fontsize=20)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
```

```python
Reg.score(x_train,y_train)
```

```python
Reg.score(x_test,y_test)
```
