```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
Data = pd.read_csv("3.2 Countries_exercise.csv.csv")
Data.head()
```

```python
Data.shape
```

```python
plt.scatter(Data["Longitude"],Data["Latitude"])
plt.xlabel("Longitude",fontsize=20)
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.ylabel("Latitude",fontsize=20)
plt.legend()
plt.show()
```

```python
X = Data.iloc[:, 1:3]
X
```

```python
KM = KMeans(3)
```

```python
KM.fit(X)
```

```python
Identified_Clusters = KM.fit_predict(X)
Identified_Clusters
```

```python
Data["Cluster"] = Identified_Clusters
Data.head()
```

```python
plt.scatter(Data["Longitude"],Data["Latitude"],c=Data["Cluster"],cmap="rainbow")
plt.xlabel("Longitude",fontsize=20)
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.ylabel("Latitude",fontsize=20)
plt.legend()
plt.show()
```

```python
KM.inertia_
```

```python
WCSS = []
for i in range(1,10):
    KM = KMeans(i)
    KM.fit(X)
    Wcss = KM.inertia_
    WCSS.append(Wcss)
```

```python
WCSS
```

```python
Cluster = range(1,10)
Cluster
```

```python
plt.plot(Cluster,WCSS)
plt.xlabel("CLUSTER",fontsize=20)
plt.ylabel("WCSS",fontsize=20)
plt.legend()
plt.show()
```
